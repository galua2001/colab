```python
D=xx.shape[1]
w=np.ones(D)
```

```python
print(w)
```

```python
def pred(xx,w):
  return( xx @ w)
```

```python
def sigmoid(x):
    return 1/(1+ np.exp(-x))
```

```python
def cross_entropy(yt, yp):
    # 교차 엔트로피의 계산 (이 단계에서는 벡터)
    ce1 = -(yt * np.log(yp) + (1 - yt) * np.log(1 - yp))
    # 교차 엔트로피 벡터의 평균값을 계산
    return(np.mean(ce1))
```

```python
# 반복 횟수
iters = 1000

# 학습률
alpha = 0.1
```

```python
for k in range(iters):

    # 예측값 계산 (7.8.1)
    yp = sigmoid(pred(xx, w))

    # 오차 계산 (7.8.2)
    yd = yp - yt

    # 경사하강법 적용 (7.8.4)
    w = w - alpha * ((xx.T).dot(yd)) /xx.shape[0]
    # 학습 곡선을 그리기 위한 데이터 계산 및 저장
    if (k % 10 == 0):
        # 손실함숫값의 계산 (7.6.1)
        loss1= cross_entropy(yt, yp)
        print(k,loss1)
```

```python
from sklearn.metrics import accuracy_score
def evaluate(xt, yt, w):

    # 예측값 계산
    sigmoid( pred(xt, w))

    # 손실함수 값 계산
    loss = cross_entropy(yt, yp)

    # 예측값(확률값)을 0이나 1로 변환
    yp_b = classify(yp)

    # 정확도 산출
    score = accuracy_score(yt, yp_b)
    return loss, score
```
