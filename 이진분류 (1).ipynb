```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
x_org, y_org=iris.data, iris.target
print(x_org, x_org.shape, y_org.shape, y_org)
```

```python
x_data, y_data=iris.data[:100,:2], iris.target[:100]
x_data=np.insert(x_data,0,1.0, axis=1)
print(x_data.shape)
```

```python
print(x_data[:100], y_data[:100])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_data, y_data, train_size=70, test_size=30, random_state=123)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
x=x_train
yt=y_train
print(x[:5])
print(yt[:5])
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def pred(x,w):
  return sigmoid(x@w)
```

```python
M=x.shape[0]
D=x.shape[1]
iters=1000
alpha=0.01
w=np.ones(D)
print(M,D,w)
```

```python
for k in range(iters):
  yp = pred(x,w)
  yd=yp-yt
  w=w-alpha*(x.T @ yd)/M
```

```python
x_t0=x_train[y_train==0]
x_t1=x_train[y_train==1]
plt.figure(figsize=(6,6))
plt.scatter(x_t0[:,1], x_t0[:,2], marker='x',c='b', label='0(setosa)')
plt.scatter(x_t1[:,1], x_t1[:,2], marker='o',c='k', label='1(versicolor)')
xx=np.linspace(4.3, 6.9, 10)
yy=-(w[0]+w[1]*xx)/w[2]
plt.plot(xx,yy)
plt.show()
```

```python
print(w[0], w[1],w[2])
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
```

```python
model_lr.fit(x,yt)
w0=model_lr.coef_[0,0]
w1=model_lr.coef_[0,1]
w2=model_lr.coef_[0,2]
```

```python
print(w0, w1, w2)
```

```python
x_t0=x_train[y_train==0]
x_t1=x_train[y_train==1]
plt.figure(figsize=(6,6))
plt.scatter(x_t0[:,1], x_t0[:,2], marker='x',c='b', label='0(setosa)')
plt.scatter(x_t1[:,1], x_t1[:,2], marker='o',c='k', label='1(versicolor)')
xx=np.linspace(4.3, 6.9, 2)
yy=-(w0-4+w1*xx)/w2
plt.plot(xx,yy)
plt.show()
```
