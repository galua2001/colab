```python
import numpy as np
import matplotlib.pyplot as plt
eta=0.01
x=3
record_x=[]
record_y=[]
for i in range(100):
  y=3*x**2-6*x+11/3
  record_x.append(x)
  record_y.append(y)
  x=x-eta*(6*x-6)
xx=np.linspace(-2,4)
yy=3*xx**2-6*xx+11/3
plt.plot(xx,yy, c='r', linestyle='dashed')
plt.scatter(record_x, record_y)
plt.show()

```
